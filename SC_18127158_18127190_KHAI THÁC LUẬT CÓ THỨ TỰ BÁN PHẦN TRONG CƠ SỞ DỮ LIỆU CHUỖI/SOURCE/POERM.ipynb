```python
file1 = open("./drive/MyDrive/KLTN/Normal/Time/1_minsup.txt","r")
file2 = open("./drive/MyDrive/KLTN/Optimize/Time/1_minsup.txt","r")
minsup1 = file1.readline()[:-1].split(" ")
time1 = file1.readline().split(" ")
minsup1 = [int(i) for i in minsup1]
time1 = [int(i) for i in time1]
plt.plot(minsup1,time1,'rD-',label='POERM')
minsup2 = file2.readline()[:-1].split(" ")
time2 = file2.readline().split(" ")
minsup2 = [int(i) for i in minsup2]
time2 = [int(i) for i in time2]
plt.plot(minsup2,time2,'bo-',label='POERM-Op')
plt.title("OnlineRetail")
plt.ylabel('time(s)')
plt.xlabel('minsup')
plt.savefig('1_1.png')
plt.legend(loc='best')
plt.show()
file1.close()
file2.close()
```
